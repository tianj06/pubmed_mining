Pubmed query syntex:

Basic Searching

esearch.fcgi?db=<database>&term=<query>
Input: Entrez database (&db); Any Entrez text query (&term)

Output: List of UIDs matching the Entrez query

Example: Get the PubMed IDs (PMIDs) for articles about breast cancer published in Science in 2008

http://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&term=science[journal]+AND+breast+cancer+AND+2008[pdat]

```python
import requests
from pattern import web 
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle

%matplotlib inline 
```

```python
pub = 'http://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed'
term = 'optogenetics'
query_url = pub + '&term=' + term + '&retmax=10000'
#print query_url
query_result = requests.get(query_url).text
#print query_result
```

```python
resultIDs = [i.content for i in  web.Element(query_result).by_tag('Id')  ]
```

Basic Downloading

esummary.fcgi?db=<database>&id=<uid_list>
Input: List of UIDs (&id); Entrez database (&db)

Output: XML DocSums

Example: Download DocSums for these protein GIs: 6678417,9507199,28558982,28558984,28558988,28558990

http://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi?db=protein&id=6678417,9507199,28558982,28558984,28558988,28558990

```python
# download all results and save it 
dom = web.Element(query_result)
#save data locally
allxml = list()
for ID in dom.by_tag('Id'):
    ID = ID.content
    pubsum = 'http://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed'
    sum_url = pubsum + '&id=' + ID+'&retmode=xml'
    xml = requests.get(sum_url).text
    allxml.append(xml)

#save all search results 
import pickle
f = open('allxml.pckl', 'w')
pickle.dump(allxml, f)
f.close()
```

For each result, extract 1) first author 2) last author 3) first address of corresponding author 4) publish date 5) journal 6) publication type (research or review)

```python
def getpubdate(dom):   
    pubmed_date = dom.by_tag('PubMedPubDate')
    attrlist = [p.attrs['pubstatus'] for p in pubmed_date]
    if u'accepted' in attrlist:
        date = pubmed_date[attrlist.index(u'accepted')]
    else:
        date = pubmed_date[attrlist.index(u'pubmed')]
    year = date.by_tag('Year')[0].content
    month = date.by_tag('Month')[0].content
    if len(month)==1:
        month = '0' + month
    day = date.by_tag('Day')[0].content
    if len(day)==1:
        day = '0' + day
    return datetime.strptime(year + month + day, '%Y%m%d')
```

```python
def extract_name(d):
    """ This function extract author name, i.e. "AY Wang"
        input is Author tag
    """
    initials = d.by_tag('Initials')[0].content
    lastname = d.by_tag('LastName')[0].content
    return initials + ' ' + lastname

def extract_affili(d):
    """ This function extract author affiliation
        input is Author tag
        if there are mulitple affiliations, only keep the first one
    """
    af_info = d.by_tag('AffiliationInfo')
    if af_info:
        return af_info[0].by_tag('affiliation')[0].content
    else:
        return None

def get_authors(dom):
    """ This function returns all authors' name and affiliation as a list of dicts
    """
    try:
        author_list = dom.by_tag('AuthorList')[0]
        result = list()
        defaultAf = None
        for au in author_list.by_tag('Author'):
            author_name = extract_name(au)
            author_affliation = extract_affili(au)
            if (author_affliation!=None) and (defaultAf is None):
                defaultAf = author_affliation
            result.append({'AuthorName':author_name, 'Affiliation':author_affliation}) 
        # replace author without affiliatiion by the default value
        filled_result = list()
        for item in result:
            if item['Affiliation'] is None:
                item['Affiliation'] = defaultAf
            filled_result.append(item)   
    except:
        filled_result = None
    return filled_result
```

```python
def get_journal(dom):
    journal = dom.by_tag('Journal')[0]
    result = {'Title':journal.by_tag('Title')[0].content, 
              'Abbre':journal.by_tag('ISOAbbreviation')[0].content }
    return result
#print get_journal(dom) 
```

```python
def get_pubtype(dom):
    """ This function return paper publication type
        To do: check whether one article has several types
    """
    return dom.by_tag('PublicationType')[0].content 
#print get_pubtype(dom)
```

```python
def get_keyword(dom):
    return [k.content for k in dom.by_tag('Keyword')]
#print get_keyword(dom)        
```

```python
def get_all_info_per_ID(xml):
    """ This function extracts specific information given a pubmedID
    """
    dom = web.Element(xml)
    return { 'pubdate' : getpubdate(dom),
             'authors' : get_authors(dom),
             'journal' : get_journal(dom),
             'type' : get_pubtype(dom),
             'keywords' : get_keyword(dom)}
```

```python
def get_impact_factor (journalName):
    
    """ SCI websites provide impact factor for journals
        for example: http://www.scijournal.org/impact-factor-of-J-NEUROSCI.shtml
        we can construct the url by journal abbrevation
    """
    journalName = re.sub(r'[.]', '', journalName)
    journalName = re.sub(r' ', '-', journalName)
    sci_url = 'http://www.scijournal.org/impact-factor-of-' + journalName.upper() +'.shtml'
    if web.URL(sci_url).exists:       
        html = requests.get(sci_url).text
        dom = web.Element(html)
        #print dom
        d = dom('div#main_content')[0].by_tag('h2')[0]
        #print d
        impact_factors = list()
        for a in re.findall(r'Impact Factor :.\d+.\d\d\d', d.content):
            impact_factors.append(float(a.split(':')[1]))
        return np.mean(impact_factors)
    else:
        return np.nan
```
